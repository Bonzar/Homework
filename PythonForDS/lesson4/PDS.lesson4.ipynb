## task3

```python
x = np.linspace(-3, 3, 51)
# print(x)
```

```python
y1 = x*x
y2 = 2*x + 0.5
y3 = -3*x - 1.5
y4 = np.sin(x)

# print(f'y1:\n {y1}\n'
#       f'y2:\n {y2}\n'
#       f'y3:\n {y3}\n'
#       f'y4:\n {y4}')
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)

ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, y1)
ax1.set_title("График y1")
ax1.set_xlim(-5, 5)

ax2.plot(x, y2)
ax2.set_title("График y2")

ax3.plot(x, y3)
ax3.set_title("График y3")

ax4.plot(x, y4)
ax4.set_title("График y4")

fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)
```

```python
import pandas as pd
plt.style.use('fivethirtyeight')
```

```python
creditcard = pd.read_csv("creditcard.csv")
creditcard
```

```python
creditcard_Class_val_counts = pd.value_counts(creditcard["Class"])
```

```python
creditcard_Class_val_counts.plot(kind="bar")
plt.show()
```

```python
creditcard_Class_val_counts.plot(kind="bar", logy=True)
plt.show()
```

```python
creditcard_V1 = creditcard.loc[creditcard["Class"] == 1, "V1"]
creditcard_V2 = creditcard.loc[creditcard["Class"] == 0, "V2"]
```

```python
plt.hist(
         [creditcard_V1, creditcard_V2], 
         bins=20, 
         density=True, 
         alpha=0.5, 
         color=["grey", "red"], 
         label=["Class 0","Class 1"]
        )
plt.legend()
plt.xlabel("Class", fontsize=14)
```
