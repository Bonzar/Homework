```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-50, 50, 5000)
def func1(_x): return _x**2*(np.sin(0.5*_x)**2+1)
def grad1(_x): return _x*(2*np.sin(0.5 * _x)**2 + _x*np.sin(0.5*_x) * np.cos(0.5 * _x) + 2)
```

```python
plt.plot(x, func1(x))
plt.show()
```

```python
x0=np.random.uniform(-50, 50)

# шаг градиентного спуска
alpha = 0.05

# максимальное число итераций
max_iter = 4000

# критерий сходимости (разница весов, при которой алгоритм останавливается)
min_weight_dist = 1e-6

# зададим начальную разницу весов большим числом
weight_dist = np.inf

# счетчик итераций
iter_num = 0

while weight_dist > min_weight_dist and iter_num < max_iter:
    new_x0 = x0 - alpha*grad1(x0)
    weight_dist = np.linalg.norm(new_x0 - x0)
    print(x0, func1(x0))

    
    iter_num += 1
    x0 = new_x0
```
